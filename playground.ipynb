```python
feat = "moup_timelags~1"

feats = {}


for session,sdict in dict.items():
    for parcellation,pdict in sdict.items():
        for trials,tdict in pdict.items():
            for condition,cdict in tdict.items():
                feats[condition] = feats.get(condition, {})
                feats[condition][session]=cdict[feat][:,0,:,:]


n_conditions = len(feats)

n_sessions = len(list(feats.values())[0])
t,w,h =  list(list(feats.values())[0].values())[0].shape

feats_array_mean = np.zeros((n_conditions,n_sessions,w,h))
feats_array_std = np.zeros((n_conditions,n_sessions,w,h))

for c , (_, cdict) in enumerate(feats.items()):
    for s, (_, val) in enumerate(cdict.items()):
        feats_array_mean[c,s,:,:] =  np.mean(val,axis=0)
        feats_array_std[c,s,:,:] =  np.std(val,axis=0)
        
#print(feats_array)
pd.DataFrame.from_dict(feats)

print(np.mean(feats_array_mean,axis=1))
                
```
